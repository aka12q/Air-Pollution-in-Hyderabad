```python
data = pd.read_csv("AQI Data Set.csv")
data.head()
```

```python
data.rename(columns={'Mounths':'Month'},inplace=True)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
data.head()
```

```python
data['Month']=data['Month'].astype(str)
data
```

```python
new = data['Month'].str.split("-", expand = True)
print(new)
```

```python
data['Month'] = new[0]
data['Year'] = new[1]
```

```python
month={'Jan':1,'Feb':2,'Mar':3,'Apr':4,
       'May':5,'Jun':6,'Jul':7,'Aug':8,
       'Sep':9,'Oct':10,'Nov':11,'Dec':12
      }
data['Month']=data['Month'].map(month)
data.head()
```

```python
data.drop(['Id'], axis=1, inplace=True)
```

```python
data['Year'] = 2000 + data['Year'].astype(int)
data['Month'] = data['Month'].astype(int)
data.head()
```

# Correlation Analysis

```python
for col in data.columns:
    if col != 'AQI' and  col!="Month":
        sns.regplot(x=col, y='AQI', data=data)
        c = data['AQI'].corr(data[col])
        plt.title(f'Correlation between {col} and AQI (Correlation Coefficient: {c:.2f})')
        plt.xlabel(col)
        plt.ylabel('AQI')
        plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr()).set(title='Correlation')
```

# Mean and Median of Values

```python
Mean=[
    data['PM10 in æg/m3'].mean(),
    data['SO2 in æg/m3'].mean(),
    data['NOx  in æg/m3'].mean(),
    data[' PM2.5  in æg/m3'].mean(),
    data['Ammonia - NH3  in æg/m3'].mean(),
    data['O3   in æg/m3'].mean(),
    data['CO  in mg/m3'].mean(),
    data[' Benzene  in æg/m3'].mean(),
    data['AQI'].mean()
]
columns = ['PM10 in æg/m3', 'SO2 in æg/m3', 'NOx  in æg/m3',
       'PM2.5  in æg/m3', 'Ammonia - NH3  in æg/m3', 'O3   in æg/m3',
       'CO  in mg/m3', 'Benzene  in æg/m3', 'AQI']

print('Mean of all the column Values\n')
for i in range(len(columns)):
    print(columns[i]," ",Mean[i])
```

```python
Median=[
    data['PM10 in æg/m3'].median(),
    data['SO2 in æg/m3'].median(),
    data['NOx  in æg/m3'].median(),
    data[' PM2.5  in æg/m3'].median(),
    data['Ammonia - NH3  in æg/m3'].median(),
    data['O3   in æg/m3'].median(),
    data['CO  in mg/m3'].median(),
    data[' Benzene  in æg/m3'].median(),
    data['AQI'].median()
]
columns = ['PM10 in æg/m3', 'SO2 in æg/m3', 'NOx  in æg/m3',
       'PM2.5  in æg/m3', 'Ammonia - NH3  in æg/m3', 'O3   in æg/m3',
       'CO  in mg/m3', 'Benzene  in æg/m3', 'AQI']

print('Median of all the column Values\n')
for i in range(len(columns)):
    print(columns[i]," ",Median[i])
```
